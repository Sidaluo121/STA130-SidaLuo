```python
import numpy as np

# Sample data (you can replace this with your actual data)
original_data = np.array([5, 7, 8, 6, 9])

# Number of bootstrap samples
n_bootstraps = 1000

# Initialize a list to hold the bootstrapped means
bootstrapped_means = []

# Generate bootstrapped samples and calculate means
for _ in range(n_bootstraps):
    # Draw a bootstrap sample with replacement
    bootstrap_sample = np.random.choice(original_data, size=len(original_data), replace=True)
    
    # Calculate the mean of the bootstrap sample
    bootstrapped_means.append(np.mean(bootstrap_sample))

# Convert the list to a NumPy array for easier manipulation
bootstrapped_means = np.array(bootstrapped_means)

# Calculate the 2.5th and 97.5th percentiles for the confidence interval
lower_bound = np.quantile(bootstrapped_means, 0.025)  # 2.5th percentile
upper_bound = np.quantile(bootstrapped_means, 0.975)  # 97.5th percentile

# Print the 95% confidence interval for the mean
print("95% Confidence Interval for the Mean:", (lower_bound, upper_bound))

# ---- Modifying the Code for Different Population Parameters ----

# To calculate the 95% bootstrap confidence interval for the median instead of the mean,
# simply replace the calculation of the mean with the calculation of the median.
bootstrapped_medians = []

# Generate bootstrapped samples and calculate medians
for _ in range(n_bootstraps):
    bootstrap_sample = np.random.choice(original_data, size=len(original_data), replace=True)
    
    # Calculate the median of the bootstrap sample instead of the mean
    bootstrapped_medians.append(np.median(bootstrap_sample))

# Convert the list to a NumPy array
bootstrapped_medians = np.array(bootstrapped_medians)

# Calculate the 2.5th and 97.5th percentiles for the confidence interval for the median
lower_bound_median = np.quantile(bootstrapped_medians, 0.025)  # 2.5th percentile
upper_bound_median = np.quantile(bootstrapped_medians, 0.975)  # 97.5th percentile

# Print the 95% confidence interval for the median
print("95% Confidence Interval for the Median:", (lower_bound_median, upper_bound_median))
```
