```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv'
data = pd.read_csv(url)

# Overall summary
print("Overall summary:")
print(data.info())

# Summary of numerical columns
print("\nSummary of numerical columns:")
print(data.describe())

# Summary of categorical columns
print("\nSummary of categorical columns:")
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column}:")
    print(data[column].value_counts())

# Check for missing values
print("\nMissing values:")
print(data.isnull().sum())
```

An attribute is a variable that belongs to an object, and they are access using dot notation; a method represents the behavior or actions that an object can perform, and it is basically a function
