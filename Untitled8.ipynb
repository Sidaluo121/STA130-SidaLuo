Q1:
    1. Simple linear regression models the relationship between one predictor and one outcome variable, using equations like Y = beta0 + beta1X + epsilon. Multiple linear regression extends this by including multiple Xs and Ys like Y = beta0 + beta1X1 + beta2X2 and so on. Multiple linear regression allows for the analysis of the effect of each predictor while controlling for others, providing a more comprehensive understamding of the relationships between multiple variables.
    2. A continuous varaible can take an infinite number of values, so can be modeled as Y = beta0 + beta1X. While an indicator varaible is binary, and indicates one of two types of one category. It can be modeled as Y = beta0 + beta1 * indicator.
    3. Adding an indicator varaible would make the model to differentiate groups based on the indicator while accounting for the continuous varaible's effect.
    4. Adding an interaction between continuous and indicator variables would allow the effect of the continuous varaible to vary by the category defined by the indicator variable. This allows modeling situations where the relationship between the continuous variable and the outcome varaible depend on the indicator variable.
    5. When a non-binary categorical varaible is used, each category except one is represented by an indicator variable. The excluded category serves as the baseline. This approach allows the model to represent different group means for each category, relative to the baseline.

```python
import plotly.express as px
import numpy as np
import statsmodels.formula.api as smf

# Convert gender to binary for interaction purposes
cscs_data['gender_binary'] = (cscs_data['DEMO_gender'] == 'Man').astype(int)

# Define and fit the additive model
additive_formula = 'outcome_binary ~ DEMO_age + gender_binary'
log_reg_fit_additive = smf.logit(additive_formula, data=cscs_data).fit()

# Define and fit the synergistic model with interaction between age and gender
synergistic_formula = 'outcome_binary ~ DEMO_age * gender_binary'
log_reg_fit_synergistic = smf.logit(synergistic_formula, data=cscs_data).fit()

# Simulate values for age for plotting
x_vals = np.linspace(cscs_data['DEMO_age'].min(), cscs_data['DEMO_age'].max(), 100)

# Compute y-values for the additive model
y_vals_additive = (
    log_reg_fit_additive.params['Intercept'] +
    log_reg_fit_additive.params['DEMO_age'] * x_vals +
    log_reg_fit_additive.params['gender_binary'] * cscs_data['gender_binary'].mean()
)

# Plot for additive specification
fig_additive = px.scatter(x=x_vals, y=y_vals_additive, labels={'x': 'Age', 'y': 'Predicted Outcome'})
fig_additive.update_traces(mode='lines+markers')
fig_additive.show(renderer = 'png')

# Compute y-values for the synergistic model with interaction
y_vals_synergistic = (
    log_reg_fit_synergistic.params['Intercept'] +
    log_reg_fit_synergistic.params['DEMO_age'] * x_vals +
    log_reg_fit_synergistic.params['gender_binary'] * cscs_data['gender_binary'].mean() +
    log_reg_fit_synergistic.params['DEMO_age:gender_binary'] * x_vals * cscs_data['gender_binary'].mean()
)

# Plot for synergistic specification
fig_synergistic = px.scatter(x=x_vals, y=y_vals_synergistic, labels={'x': 'Age', 'y': 'Predicted Outcome'})
fig_synergistic.update_traces(mode='lines+markers')
fig_synergistic.show(renderer = 'png')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

The first cell handles data preprocessing and splitting for training and testing. It fills missing values in the Type 2 column. The dataset is then split 50/50 into training and testing sets.
The second cell defined a simple linear regression model to predict HP using only attack and defense as predictors. It fits the model on the training data.
The third cell calculates both in-sample and out-of-sample R^ values.
The fourth cell specified Model 4 with multiple interactions among a number of varaibles.
The fifth cell calculates R^ for model 4.

```python
# "Cond. No." WAS 343.0 WITHOUT to centering and scaling
model3_fit.summary() 
from patsy import center, scale

model3_linear_form_center_scale = \
  'HP ~ scale(center(Attack)) + scale(center(Defense))' 
model_spec3_center_scale = smf.ols(formula=model3_linear_form_center_scale,
                                   data=pokeaman_train)
model3_center_scale_fit = model_spec3_center_scale.fit()
model3_center_scale_fit.summary()
# "Cond. No." is NOW 1.66 due to centering and scaling
model4_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Defense))'
model4_linear_form_CS += ' * scale(center(Speed)) * Legendary' 
model4_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
# Legendary is an indicator, so we don't center and scale that

model4_CS_spec = smf.ols(formula=model4_linear_form_CS, data=pokeaman_train)
model4_CS_fit = model4_CS_spec.fit()
model4_CS_fit.summary().tables[-1]  # Cond. No. is 2,250,000,000,000,000

# The condition number is still bad even after centering and scaling
# Just as the condition number was very bad to start with
model4_fit.summary().tables[-1]  # Cond. No. is 12,000,000,000,000,000
```

In model4, the complex interactions among predictors create a highly multicollinear design matrix, where multiple columns (predictors and interaction terms) are strongly correlated. This multicollinearity inflates the condition number of the design matrix, signaling potential overfitting. Due to these correlations, the model learns specific patterns unique to the training data (often noise) rather than generalizable relationships. Consequently, while model4 may perform well in-sample, it generalizes poorly out-of-sample, illustrating how excessive model complexity can reduce predictive reliability on new data.

```python
# Here's something a little more reasonable...
model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary'
model5_linear_form += ' + Q("Sp. Def") + Q("Sp. Atk")'
model5_linear_form += ' + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

model5_spec = smf.ols(formula=model5_linear_form, data=pokeaman_train)
model5_fit = model5_spec.fit()
model5_fit.summary()
yhat_model5 = model5_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model5_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model5)[0,1]**2)
# Here's something a little more reasonable...
model6_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
# And here we'll add the significant indicators from the previous model
# https://chatgpt.com/share/81ab88df-4f07-49f9-a44a-de0cfd89c67c
model6_linear_form += ' + I(Q("Type 1")=="Normal")'
model6_linear_form += ' + I(Q("Type 1")=="Water")'
model6_linear_form += ' + I(Generation==2)'
model6_linear_form += ' + I(Generation==5)'

model6_spec = smf.ols(formula=model6_linear_form, data=pokeaman_train)
model6_fit = model6_spec.fit()
model6_fit.summary()
yhat_model6 = model6_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model6_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model6)[0,1]**2)
# And here's a slight change that seems to perhaps improve prediction...
model7_linear_form = 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk")'
model7_linear_form += ' + I(Q("Type 1")=="Normal")'
model7_linear_form += ' + I(Q("Type 1")=="Water")'
model7_linear_form += ' + I(Generation==2)'
model7_linear_form += ' + I(Generation==5)'

model7_spec = smf.ols(formula=model7_linear_form, data=pokeaman_train)
model7_fit = model7_spec.fit()
model7_fit.summary()
yhat_model7 = model7_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model7_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model7)[0,1]**2)
# And here's a slight change that seems to perhas improve prediction...
model7_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Speed))'
model7_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
# We DO NOT center and scale indicator variables
model7_linear_form_CS += ' + I(Q("Type 1")=="Normal")'
model7_linear_form_CS += ' + I(Q("Type 1")=="Water")'
model7_linear_form_CS += ' + I(Generation==2)'
model7_linear_form_CS += ' + I(Generation==5)'

model7_CS_spec = smf.ols(formula=model7_linear_form_CS, data=pokeaman_train)
model7_CS_fit = model7_CS_spec.fit()
model7_CS_fit.summary().tables[-1] 
# "Cond. No." is NOW 15.4 due to centering and scaling
# "Cond. No." WAS 2,340,000,000 WITHOUT to centering and scaling
model7_fit.summary().tables[-1]
```

The progression from model3 to model7 reflects an effort to improve predictive accuracy while balancing model complexity and generalizability:

model5 expands on model3 by adding relevant predictors (e.g., Speed, Sp. Def, Generation categories), which increases complexity but risks multicollinearity and overfitting.
model6 simplifies model5 by retaining only significant predictors, reducing redundant variables to enhance generalizability.
model7 builds on model6 by introducing interactions (e.g., Attack * Speed * Sp. Def * Sp. Atk) to capture complex relationships while managing predictor count.
Multicollinearity Check: Centering and scaling model7's continuous predictors yields a condition number of 15.4, indicating that multicollinearity is not a major concern.
In summary, each model extension seeks to capture meaningful relationships in a way that generalizes well, using a balance of evidence-based variable selection and multicollinearity diagnostics.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split

# Parameters for loop
reps = 100
in_sample_Rsquared = np.zeros(reps)
out_of_sample_Rsquared = np.zeros(reps)

# Linear form (you may choose a different model specification, e.g., model3)
linear_form = 'HP ~ Attack + Defense'

# Loop to calculate in-sample and out-of-sample R^2 across multiple random splits
for i in range(reps):
    # Random 50-50 split for each iteration
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Fit the model on the training data
    model_fit = smf.ols(formula=linear_form, data=pokeaman_train).fit()
    
    # Collect in-sample R-squared
    in_sample_Rsquared[i] = model_fit.rsquared
    
    # Collect out-of-sample R-squared
    y_test = pokeaman_test['HP']
    y_pred = model_fit.predict(pokeaman_test)
    out_of_sample_Rsquared[i] = np.corrcoef(y_test, y_pred)[0, 1]**2

# Create a DataFrame for visualization
df = pd.DataFrame({
    "In Sample Performance (R-squared)": in_sample_Rsquared,
    "Out of Sample Performance (R-squared)": out_of_sample_Rsquared
})

# Scatter plot of in-sample vs out-of-sample R-squared
fig = px.scatter(df, x="In Sample Performance (R-squared)", y="Out of Sample Performance (R-squared)",
                 title="In-Sample vs Out-of-Sample R-squared across Random Splits")
# Add y=x line to visualize ideal generalizability
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y=x", line_shape='linear', line=dict(dash='dash')))
fig.show(renderer = 'png')
```
