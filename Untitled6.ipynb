```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Define theoretical parameters
n_samples = 100
X = np.linspace(0, 10, n_samples)  # Predictor variable
true_slope = 2.5
true_intercept = 5

# Step 2: Generate normally distributed errors
error = np.random.normal(loc=0, scale=2, size=n_samples)

# Step 3: Generate the outcome variable (Y) based on the theoretical model
Y = true_intercept + true_slope * X + error  # Y = 0 + 1X + 

# Step 4: Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 5: Fit the Simple Linear Regression model
model = smf.ols('Y ~ X', data=data).fit()

# Step 6: Generate predictions
data['predicted_Y'] = model.predict(data)

# Step 7: Visualize the fitted model and theoretical line
plt.figure(figsize=(10, 6))

# Plot the original data points
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue', alpha=0.6)

# Plot the fitted regression line
plt.plot(data['X'], data['predicted_Y'], color='red', label='Fitted Regression Line', linewidth=2)

# Plot the theoretical regression line
theoretical_Y = true_intercept + true_slope * X
plt.plot(X, theoretical_Y, color='green', linestyle='--', label='Theoretical Regression Line', linewidth=2)

# Add labels and title
plt.xlabel('Predictor Variable (X)')
plt.ylabel('Outcome Variable (Y)')
plt.title('Fitted Simple Linear Regression Model with Theoretical Line')
plt.legend()
plt.show()

# Optional: Print the model summary
print(model.summary())
```

```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n,x_min,x_range,beta0,beta1,sigma = 20,5,5,2,3,5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
y = beta0 + beta1 * x + errors

df = pd.DataFrame({'x': x, 'y': y})
model_data_specification = smf.ols("Y~x", data=df) 
fitted_model = model_data_specification.fit() 

df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
```

OLS finds the line that best captures the underlying trend in the data by minimizing the square difference between observed and predicted values, balancing the accuracy and stability in the model's predictions.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import plotly.express as px

# Load dataset and filter for long wait times
old_faithful = sns.load_dataset('geyser')
long_wait_threshold = 70
long_wait = old_faithful[old_faithful['waiting'] > long_wait_threshold]

# Parameters for null hypothesis (no association)
beta0 = 1.65
beta1 = 0
theta = 0.37
n_bootstraps = 1000  # Number of bootstrap samples

# Step 1: Bootstrap sampling distribution of the slope
bootstrap_slopes = []
for _ in range(n_bootstraps):
    # Bootstrap sample
    bootstrap_sample = long_wait.sample(n=160, replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])  # Collect slope

# Step 2: Null hypothesis simulation (no association)
null_slopes = []
for _ in range(n_bootstraps):
    # Simulate duration data under the null hypothesis (beta1=0)
    simulated_duration = beta0 + beta1 * long_wait['waiting'] + np.random.normal(0, theta, len(long_wait))
    simulated_data = pd.DataFrame({'waiting': long_wait['waiting'], 'duration': simulated_duration})
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    null_slopes.append(model.params['waiting'])  # Collect slope

# Step 3: Visualization of bootstrap and null distributions
# Create DataFrames for Plotly visualization
bootstrap_df = pd.DataFrame({'Slope': bootstrap_slopes, 'Type': 'Bootstrap'})
null_df = pd.DataFrame({'Slope': null_slopes, 'Type': 'Null Hypothesis'})
combined_df = pd.concat([bootstrap_df, null_df])

fig = px.histogram(combined_df, x='Slope', color='Type', nbins=30, barmode='overlay',
                   title="Bootstrap and Null Hypothesis Sampling Distributions of Slope Coefficients")
fig.show(renderer="png")

# Step 4: Compute 95% Bootstrap Confidence Interval for the Slope
bootstrap_ci = np.percentile(bootstrap_slopes, [2.5, 97.5])
print("95% Bootstrap Confidence Interval for Slope:", bootstrap_ci)

# Step 5: Check if 0 is within the 95% CI
contains_zero = (bootstrap_ci[0] <= 0 <= bootstrap_ci[1])
print("Does the 95% CI contain 0?", contains_zero)

# Step 6: Calculate p-value from null hypothesis simulations
# p-value is the proportion of null hypothesis slopes that are as extreme as the observed slope
observed_slope = smf.ols('duration ~ waiting', data=long_wait).fit().params['waiting']
p_value = (np.sum(np.abs(null_slopes) >= np.abs(observed_slope)) / n_bootstraps)
print("Simulated p-value:", p_value)

# Step 7: Compare with p-value from original model fit
original_model_p_value = smf.ols('duration ~ waiting', data=long_wait).fit().pvalues['waiting']
print("Original model p-value:", original_model_p_value)
```
